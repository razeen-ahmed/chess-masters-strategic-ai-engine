```python
import random
import math



def strength(x):
    val = math.log2(x+1)+(x/10)
    # print("strength", val)
    return val


def utility_val(maxV,minV):

    i = random.randint(0,1)
    val = strength(maxV) - strength(minV) + ((-1)**(i))*(random.randint(1,10)/10)
    # print(val)
    return val



#Chess Minimax Algorithm
def chess_minmax(game_depth, alpha, beta, maximize, val1, val2):

    if game_depth == 0:
        calc = utility_val(val1,val2)
        return round(calc,2)

    if maximize == True:

        max_node = -math.inf

        for i in range(2): #creating branch node branching factor = 2
            value = chess_minmax(game_depth-1, alpha, beta, False, val1, val2)

            max_node = max(max_node,value)
            alpha = max(alpha,value)

            if alpha >=beta:
                break

        return max_node

    else:
        min_node = math.inf
        for i in range(2):
            value = chess_minmax(game_depth-1, alpha, beta, True, val1, val2)

            min_node = min(min_node,value)
            beta = min(beta,value)

            if alpha >=beta:
                break

        return min_node



#All the 4 chess matches are executing here -- >
print()
def chess_finals(chess_deep, alpha, beta, magnus_strength, fabiano_strength, starting_player):

    magnus_wins = 0
    fabiano_wins = 0
    draw = 0

    for i in range(4):

        if (starting_player == 0 and i%2==0): #Magnus starts as Max Player

            max_val = magnus_strength
            min_val = fabiano_strength

            result = chess_minmax(chess_deep, alpha, beta, True, max_val, min_val)

            if result>0:
                magnus_wins +=1
                print(f"Game {i} Winner: Carlsen (Max) (Utility value: {result})")
            elif result<0:
                fabiano_wins +=1
                print(f"Game {i} Winner: Caruana (Min) (Utility value: {result})")

            else:
                draw+=1

        elif (starting_player == 0 and i%2!=0):

            max_val = fabiano_strength
            min_val = magnus_strength

            result = chess_minmax(chess_deep, alpha, beta, True, max_val, min_val)
            if result<0:
                magnus_wins +=1
                print(f"Game {i} Winner: Carlsen (Min) (Utility value: {result})")
            elif result>0:
                fabiano_wins +=1
                print(f"Game {i} Winner: Caruana (Max) (Utility value: {result})")

            else:
                draw+=1

        if (starting_player == 1 and i%2==0): #Fabiano starts as Max Player

            max_val = fabiano_strength
            min_val = magnus_strength

            result = chess_minmax(chess_deep, alpha, beta, True, max_val, min_val)

            if result>0:
                fabiano_wins +=1
                print(f"Game {i} Winner: Caruana (Max) (Utility value: {result})")
            elif result<0:
                magnus_wins +=1
                print(f"Game {i} Winner: Carlsen (Min) (Utility value: {result})")

            else:
                draw+=1

        elif (starting_player == 1 and i%2!=0):

            max_val = magnus_strength
            min_val = fabiano_strength

            result = chess_minmax(chess_deep, alpha, beta, True, max_val, min_val)

            if result>0:
                magnus_wins +=1

                print(f"Game {i} Winner: Carlsen (Max) (Utility value: {result})")
            elif result<0:
                fabiano_wins +=1
                print(f"Game {i} Winner: Caruana (Min) (Utility value: {result})")

            else:
                draw+=1


    return magnus_wins, fabiano_wins, draw


#Taking all the inputs-->
starting_player = int(input("Enter starting player for game 1 (0 for Carlsen, 1 for Caruana): "))
magnus_strength = float(input("Enter base strength for Carlsen: "))
fabiano_strength = float(input("Enter base strength for Caruana: "))


#Initiating The Chess Championship Finals!
magnus, fabiano, draw = chess_finals(5, -math.inf, math.inf, magnus_strength, fabiano_strength, starting_player)


#Overall Outputs--->

print()
print("Overall Results:")

print(f"Magnus Carlsen Wins: {magnus}")
print(f"Fabiano Caruana Wins: {fabiano}")
print(f"Draws: {draw}")
if magnus>fabiano:
    print("Overall Winner: Magnus Carlsen")
elif fabiano>magnus:
    print("Overall Winner: Fabiano Caruana")
else:
    print("Overall Winner: Draw")


```

```python
import random
import math


def strength(x):
    val = math.log2(x+1)+(x/10)
    # print("strength", val)
    return val


def utility_val(maxV,minV):

    i = random.randint(0,1)
    val = strength(maxV) - strength(minV) + ((-1)**(i))*(random.randint(1,10)/10)
    # print(val)
    return val



#Light Vs L ultimate Chess Minimax Algorithm
def deathnote_minmax(game_depth, alpha, beta, maximize, val1, val2):

    if game_depth == 0:
        calc = utility_val(val1,val2)
        return round(calc,2)

    if maximize == True:

        max_node = -math.inf

        for i in range(2): #creating branch node branching factor = 2
            value = deathnote_minmax(game_depth-1, alpha, beta, False, val1, val2)

            max_node = max(max_node,value)
            alpha = max(alpha,value)

            if alpha >=beta:
                break

        return max_node

    else:
        min_node = math.inf

        for i in range(2):
            value = deathnote_minmax(game_depth-1, alpha, beta, True, val1, val2)

            min_node = min(min_node,value)
            beta = min(beta,value)

            if alpha >=beta:
                break

        return min_node

#Light Vs L ultimate mindcontrol Chess Minimax Algorithm
def deathnote_mindcontrol_minmax(game_depth, alpha, beta, maximize, val1, val2):

    if game_depth == 0:
        calc = utility_val(val1,val2)
        return round(calc,2)

    if maximize == True:

        max_node = -math.inf

        for i in range(2): #creating branch node branching factor = 2
            value = deathnote_mindcontrol_minmax(game_depth-1, alpha, beta, True, val1, val2)

            max_node = max(max_node,value)
            alpha = max(alpha,value)

            if alpha >=beta:
                break

        return max_node

#Taking all the inputs-->

starting_player = int(input("Enter who goes first (0 for Light, 1 for L): "))
mind_control = float(input("Enter the cost of using Mind Control: "))
l_strength = float(input("Enter base strength for Light: "))
light_strength = float(input("Enter base strength for L: "))

max_val = 0
min_val = 0

deathnote_utility = 0

deathnote_mindcontrol_utility = 0

op_player = None

if starting_player == 0:
    max_val = light_strength
    min_val = l_strength

    op_player = "Light"

    deathnote_utility = deathnote_minmax(5, -math.inf, math.inf, True, max_val, min_val)

    mindcontrol_utility = deathnote_mindcontrol_minmax(5, -math.inf, math.inf, True, max_val, min_val)

elif starting_player == 1:
    max_val = l_strength
    min_val = light_strength

    op_player = "L"

    deathnote_utility = deathnote_minmax(5, -math.inf, math.inf, True, max_val, min_val)

    mindcontrol_utility = deathnote_mindcontrol_minmax(5, -math.inf, math.inf, True, max_val, min_val)


#Overall Results---->
print()
print(f"Minimax value without Mind Control: {deathnote_utility}")
print(f"Minimax value with Mind Control: {mindcontrol_utility}")
print(f"Minimax value with Mind Control after incurring the cost: {round(mindcontrol_utility-mind_control,2)}")


print()
#All cases when should one use mind control
if (deathnote_utility>0) and (mindcontrol_utility-mind_control>0):

    print(f"{op_player} should NOT use Mind Control as the position is already winning.")

elif (deathnote_utility<=0) and (mindcontrol_utility-mind_control>0):

    print(f"{op_player} should use Mind Control.")

elif (deathnote_utility<=0) and (mindcontrol_utility-mind_control<=0):

    print(f"{op_player} should NOT use Mind Control as the position is losing either way.")

elif (deathnote_utility>0) and (mindcontrol_utility-mind_control<=0):

    print(f"{op_player} should NOT use Mind Control as it backfires.")

```
